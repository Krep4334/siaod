### Задание №3
Реализовать методы сортировки строк числовой матрицы в соответствии с
заданием. Оценить время работы каждого алгоритма сортировки и сравнить его со
временем стандартной функции сортировки. Испытания проводить на сгенерированных
матрицах.

```python
import time
import numpy as np
```

```python
# Сортировка выбором.

start_time = time.time()

def selection_sort(arr):
    for i in range(len(arr)):
        min_idx = i
        for j in range(i + 1, len(arr)):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]

for row in matrix:
    selection_sort(row)
    print(row)
selection_sort_time = round((time.time() - start_time) * 1000)
print("Selection Sort --- {0} ms ---".format(selection_sort_time))
```

```python
# Сортировка вставкой
start_time = time.time()

def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key

for row in matrix:
    insertion_sort(row)
    print(row)
insertion_sort_time = round((time.time() - start_time) * 1000)
print("Insertion Sort --- {0} ms ---".format(insertion_sort_time))
```

```python
# Сортировка обменом. || Сортировка пузырьком.

start_time = time.time()

def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        swapped = False
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True
        if not swapped:
            break

for row in matrix:
    bubble_sort(row)   
    print(row)
bubble_sort_time = round((time.time() - start_time) * 1000)
print("Bubble Sort --- {0} ms ---".format(bubble_sort_time))

```

```python
# Сортировка Шелла
start_time = time.time()

def shell_sort(arr):
    n = len(arr)
    gap = n // 2
    while gap > 0:
        for i in range(gap, n):
            temp = arr[i]
            j = i
            while j >= gap and arr[j - gap] > temp:
                arr[j] = arr[j - gap]
                j -= gap
            arr[j] = temp
        gap //= 2

for row in matrix:
    shell_sort(row)
    print(row)
shell_sort_time = round((time.time() - start_time) * 1000)
print("Shell Sort --- {0} ms ---".format(shell_sort_time))
```

```python
# Быстрая сортировка
start_time = time.time()

def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quick_sort(left) + middle + quick_sort(right)

for i, row in enumerate(matrix):
    matrix[i] = quick_sort(row)

for row in matrix:
    print(row)
quick_sort_time = round((time.time() - start_time) * 1000)
print("Quick Sort --- {0} ms ---".format(quick_sort_time))
```

```python
# Турнирная сортировка
start_time = time.time()

def tournament_sort(arr):
    import heapq
    heap = list(arr)
    heapq.heapify(heap)
    return [heapq.heappop(heap) for _ in range(len(arr))]

for i, row in enumerate(matrix):
    matrix[i] = tournament_sort(row)

for row in matrix:
    print(row)
tournament_sort_time = round((time.time() - start_time) * 1000)
print("Tournament Sort --- {0} ms ---".format(tournament_sort_time))
```
